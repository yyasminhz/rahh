# **Exploring the Relationship Between Depression Severity (PHQ-9 scores) and Physiological Biomarkers (SLD010H, BPXPLS, LBXCRP) in Adults: a NHANES 2007–2018 Study by Gender and Ethnicity**

# **Summary**


This study aims to explore the relationship between depression severity (measured using the Patient Health Questionnaire-9 aka PHQ-9 ) and key physiological biomarkers including sleep duration (SLD010H), resting heart rate (BPXPLS), and C-reactive protein levels (LBXCRP) using data from the NHANES 2007–2018 cycles. In addition to examining these biomarkers, the study investigates the role of several independent variables—namely age, gender, ethnicity, smoking status, physical activity level, haemoglobin, and total cholesterol—as potential predictors of depression severity.

The analysis will be stratified by gender and ethnicity to identify potential disparities in the biological and behavioral factors associated with depressive symptoms across diverse populations. By integrating both physiological and demographic variables, this study seeks to provide a more comprehensive understanding of the biological and lifestyle correlates of depression in U.S. adults.

# **Background and rationale**

In the United States, depression affects millions of adults annually, with disparities observed across gender and ethnic groups.

While psychological and environmental factors in depression are well established, growing evidence suggests that biological markers may also play a role. Exploring these physiological and demographic correlates is essential to better understand the underlying complexity of depression and how it manifests in different populations.

Identifying potential physiological patterns linked to depressive severity can contribute to the development of early warning systems, especially for individuals who may not have access to regular mental health screenings. Additionally, by examining these relationships across gender and ethnic subgroups, the study supports efforts toward more equitable and culturally sensitive healthcare practices. In the long term, these findings may inform public health strategies aimed at integrating physical and mental health care, and guide research into biomarker-based screening or prevention tools.

## **5.2 Justification of Novelty**  
Depression has been increasingly recognized as a condition with physiological components beyond psychological symptoms. Prior research has linked elevated levels of C-reactive protein (CRP), an inflammatory marker, with depression severity. For example, Shin et al. (2016) analysed NHANES 2007–2012 data and found that individuals with depression had significantly higher CRP levels, even after adjusting for confounders. Similarly, Howren, Lamkin & Suls (2009) conducted a meta-analysis of 24 studies and reported significant associations between depression and inflammatory markers, including CRP (r = 0.12, p < 0.001). reinforcing the role of systemic inflammation in depressive disorders.


However, most of these studies focused either on a single biomarker, such as CRP, or a binary classification of depression (i.e., depressed vs. non-depressed), rather than a continuous score. Additionally, the role of blood components (HGB) and cardiovascular indicators (pulse rate) in depression has received limited attention. Moreover, existing studies often fail to explore how these relationships differ by gender and ethnicity, despite known demographic disparities in both depression prevalence and biomarker levels.  





## **5.3 How this Project Fills Knowledge Gaps**  
This study builds on the existing body of research by:  
1. Using PHQ-9 as a continuous variable to better capture gradations in depression severity;

3. Stratifying results by gender and ethnicity, thus allowing for more nuanced and equitable public health insights;

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years  = ["2007","2009","2011","2013","2015","2017"]
cycles = ["E","F","G","H","I","J"]
cycle_map = {"E":5, "F":6, "G":7, "H":8, "I":9, "J":10}

# Helper function to load XPT with fallback extensions
def load_xpt(year, cycle, prefix):
    for ext in ['XPT', 'xpt']:
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.{ext}"
        try:
            df = pd.read_sas(url, format='xport', encoding='utf-8')
            return df
        except Exception:
            continue
    raise RuntimeError(f"Unable to load {prefix} file for {year} cycle {cycle}")


# -----------------------------
# Step 1: Import and filter demographics
# -----------------------------
demo_dfs = []
for year, cycle in zip(years, cycles):
    df = load_xpt(year, cycle, 'DEMO')
    # Tag with both the raw cycle code and the actual start‐year
    df['SDDSRVYR'] = cycle_map[cycle]    # 5 → 2007–08, 6 → 2009–10, etc.
    df['Year']      = int(year)          # 2007, 2009, …
    demo_dfs.append(df)

# Concatenate into one raw DEMO
df_demo_raw = pd.concat(demo_dfs, ignore_index=True)

# Now filter by age but keep those two columns
df_demo = (
    df_demo_raw
      .query('RIDAGEYR >= 18 and RIDAGEYR <= 65')
      [['SEQN','RIAGENDR','RIDRETH1','RIDAGEYR','INDFMPIR','SDDSRVYR','Year']]
)

# -----------------------------
# Step 2: Import Smoking History (SMQ020, SMQ040)
# -----------------------------
smk_dfs = [load_xpt(year, cycle, 'SMQ') for year, cycle in zip(years, cycles)]
df_smq = pd.concat(smk_dfs, ignore_index=True)[['SEQN', 'SMQ020']]

# Concatenate and select smoking variables
df_smq = pd.concat(smk_dfs, ignore_index=True)
df_smq = df_smq[['SEQN', 'SMQ020']]

# -----------------------------
# Step 3: Import cholesterol (LBXTC)
# ----------------------------
tchol_dfs = [load_xpt(year, cycle, 'TCHOL') for year, cycle in zip(years, cycles)]
df_tchol = pd.concat(tchol_dfs, ignore_index=True)[['SEQN', 'LBXTC']]

# -----------------------------
# Step 4: Import haemoglobin (LBXHGB)
# ----------------------------
hgb_dfs = [load_xpt(year, cycle, 'CBC') for year, cycle in zip(years, cycles)]
df_hgb = pd.concat(hgb_dfs, ignore_index=True)[['SEQN', 'LBXHGB']]

# -----------------------------
# Step 5: Import heart rate (BPXPLS)
# ----------------------------
hr_dfs = [load_xpt(year, cycle, 'BPX') for year, cycle in zip(years, cycles)]
df_hr = pd.concat(hr_dfs, ignore_index=True)[['SEQN', 'BPXPLS']]

# -----------------------------
# Step 6: Import sleep hours (SLD010H, SLQ300, SLQ310)
# ----------------------------

sleep_dfs = []
for year, cycle in zip(years, cycles):
    df_cycle = load_xpt(year, cycle, 'SLQ')
    if cycle in ['E','F','G','H']:
        tmp = df_cycle[['SEQN','SLD010H']].rename(columns={'SLD010H':'SleepHours'})
    else:
        tmp = df_cycle[['SEQN','SLQ300','SLQ310']].copy()
        # parse the 'HH:MM' strings into datetime
        tmp['sleep_dt'] = pd.to_datetime(tmp['SLQ300'], format='%H:%M', errors='coerce')
        tmp['wake_dt']  = pd.to_datetime(tmp['SLQ310'], format='%H:%M', errors='coerce')
        # compute hours slept, adjust negative values (overnight)
        tmp['SleepHours'] = (tmp['wake_dt'] - tmp['sleep_dt']).dt.total_seconds() / 3600
        tmp.loc[tmp['SleepHours'] < 0, 'SleepHours'] += 24
        tmp = tmp[['SEQN','SleepHours']]
    sleep_dfs.append(tmp)

df_sleep = pd.concat(sleep_dfs, ignore_index=True)
# -----------------------------
# Step 7: Import activity level (PAD680)
# ----------------------------

act_dfs = [load_xpt(year, cycle, 'PAQ') for year, cycle in zip(years, cycles)]
df_act = pd.concat(act_dfs, ignore_index=True)[['SEQN', 'PAD680']]

# Ensure PAD680 is numeric (minutes)
df_act['PAD680'] = pd.to_numeric(df_act['PAD680'], errors='coerce')

# Recode PAD680 (minutes) to hours and classify into 4 levels:
#   < 3 hrs: Very Active (Level 4)
#   3–6 hrs: Active (Level 3)
#   > 6–9 hrs: Low Activity (Level 2)
#   > 9 hrs: Sedentary (Level 1)

def recode_activity_minutes(mins):
    if pd.isna(mins):
        return None
    hours = mins / 60.0
    if hours < 3:
        return 'Very Active'
    elif 3 <= hours <= 6:
        return 'Active'
    elif 6 < hours <= 9:
        return 'Low Activity'
    else:
        return 'Sedentary'

df_act['ActivityLevel'] = df_act['PAD680'].apply(recode_activity_minutes)


# -----------------------------
# Step 8: Import PHQ9
# ----------------------------

dpq_dfs = [load_xpt(year, cycle, 'DPQ') for year, cycle in zip(years, cycles)]
df_dpq = pd.concat(dpq_dfs, ignore_index=True)
# Define PHQ-9 items
dpq_cols = [f"DPQ0{i:02d}" for i in range(10, 100, 10)]
# Replace missing codes (7,9) with NaN
for col in dpq_cols:
    df_dpq[col] = pd.to_numeric(df_dpq[col], errors='coerce').replace({7: np.nan, 9: np.nan})
# Calculate total PHQ-9 score
df_dpq['PHQ9'] = df_dpq[dpq_cols].sum(axis=1, min_count=1)
df_phq9 = df_dpq[['SEQN', 'PHQ9']]


# -----------------------------
# Step 9: Import CRP (LBXCRP, LBXHSCRP)
# ----------------------------

crp_frames = []
for y, c in zip(years, cycles):
    if c in ['E', 'F']:
        df_c = load_xpt(y, c, 'CRP')
        var = 'LBXCRP'
    elif c in ['I', 'J']:
        df_c = load_xpt(y, c, 'HSCRP')
        var = 'LBXHSCRP'
    else:
        # skip cycles G, H (2011–2014)
        continue
    temp = df_c[['SEQN', var]].rename(columns={var: 'CRP'})
    crp_frames.append(temp)
df_crp = pd.concat(crp_frames, ignore_index=True)

# -----------------------------
# Step 10: Merge All Datasets (after filtering)
# -----------------------------

df_all = (
    df_demo
    .merge(df_tchol, on='SEQN', how='left')
    .merge(df_smq,  on='SEQN', how='left')
    .merge(df_hgb,  on='SEQN', how='left')
    .merge(df_hr,   on='SEQN', how='left')
    .merge(df_sleep,on='SEQN', how='left')    # <— updated
    .merge(df_act[['SEQN','ActivityLevel']], on='SEQN', how='left')
    .merge(df_phq9, on='SEQN', how='left')
    .merge(df_crp,  on='SEQN', how='left')
)
df_all = df_all.sort_values('SEQN').reset_index(drop=True)

# re-attach the original survey cycle code
df_all = df_all.merge(
    df_demo_raw[['SEQN','SDDSRVYR']],
    on='SEQN', how='left'
)

# keep only the DEMO‐tagged cycle code
df_all = (
    df_all
    .rename(columns={'SDDSRVYR_x': 'SDDSRVYR'})
    .drop(columns=['SDDSRVYR_y'])
)

# Merge All Datasets Before Filtering
# Use raw demographics (no age restriction)
df_all_before = df_demo_raw[['SEQN']].drop_duplicates()
for df in [df_tchol, df_smq, df_hgb, df_hr, df_sleep, df_act[['SEQN', 'ActivityLevel']], df_phq9, df_crp]:
    df_all_before = df_all_before.merge(df, on='SEQN', how='left')


df_all = df_all.sort_values(['Year','SEQN']).reset_index(drop=True)
display(HTML('<h2 style="font-size:20px; margin-bottom:12px;">Merged Dataset Preview (by Year)</h2>'))
# omit the Year column in the preview
display(df_all.drop(columns=['Year']).head(12)) # shows the first two rows of each of the six cycles
```

```python
# Shapiro–Wilk Normality Tests for Continuous Variables
from scipy.stats import shapiro
import warnings

# ignore only the "N > 5000" Shapiro warning
warnings.filterwarnings(
    "ignore",
    message=".*scipy\\.stats\\.shapiro: For N > 5000, computed p-value may not be accurate.*"
)


cont_vars = ['RIDAGEYR', 'SleepHours', 'CRP', 'PHQ9', 'LBXTC', 'BPXPLS', 'LBXHGB']
display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Shapiro–Wilk Test Results with Normality Indicator</h2>'))
shapiro_results = {}
for var in cont_vars:
    data = df_all[var].dropna()
    if len(data) >= 3:  # minimum n for Shapiro
        stat, p = shapiro(data)
        shapiro_results[var] = {'W-statistic': stat, 'p-value': p}
    else:
        shapiro_results[var] = {'W-statistic': np.nan, 'p-value': np.nan}
results_df = pd.DataFrame(shapiro_results).T

# Add column indicating normality at alpha=0.05
results_df['NormalDist'] = results_df['p-value'] > 0.05
display(results_df)

# prepare
df_plot = (results_df
  .reset_index()
  .rename(columns={'index':'Variable'})
  .assign(negLog10p = lambda d: -np.log10(d['p-value']))
)

vars_   = df_plot['Variable']
heights = df_plot['negLog10p']
colors  = ['green' if ok else 'red' for ok in df_plot['NormalDist']]

# plot
fig, ax = plt.subplots(figsize=(8,4))
ax.bar(vars_, heights, color=colors, edgecolor='black', linewidth=0.5)

# α = 0.05 threshold line
threshold = -np.log10(0.05)  # ≈ 1.301
ax.axhline(threshold, color='gray', linestyle='--', linewidth=1, label='α = 0.05')

# labels & styling
ax.set_title('Shapiro–Wilk Test: –log₁₀(p-value) (green=normal, red=non-normal)')
ax.set_ylabel('-log₁₀(p-value)')
ax.set_xticks(vars_)
ax.set_xticklabels(vars_, rotation=45, ha='right')
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import math
import re

# Define your numerical variables
num_vars = ["RIDAGEYR", "SleepHours", "CRP", "PHQ9", "LBXTC", "BPXPLS", "LBXHGB"]

# Prepare for summaries
summaries = []

# Configure subplot grid
n_cols = 3
n_vars = len(num_vars)
n_rows = math.ceil(n_vars / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()

# Loop through variables and plot
for idx, var in enumerate(num_vars):
    ax = axes[idx]
    if var in df_all:
        data = df_all[var].dropna()

        # Apply variable-specific filters
        if var == "SleepHours":
            data = data[(data >= 0) & (data <= 24)]
        elif var == "CRP":
            data = data[(data >= 0) & (data <= 50)]

        # Skip empty data
        if data.empty:
            ax.set_visible(False)
            continue

        # Compute histogram
        counts, bin_edges = np.histogram(data, bins=40)
        max_idx = counts.argmax()
        max_count = counts[max_idx]
        max_range = (bin_edges[max_idx], bin_edges[max_idx + 1])

        # Plot on the subplot
        ax.hist(data, bins=40, edgecolor='black', color='skyblue')
        ax.set_title(f"Histogram of {var}")
        ax.set_xlabel(var)
        ax.set_ylabel("Frequency")
        ax.grid(axis='y', linestyle='--', alpha=0.5)

        # Store summary
        summaries.append(
            f"{var}: Highest frequency bin is {max_range[0]:.2f} to {max_range[1]:.2f} with {max_count} observations."
        )
    else:
        ax.set_visible(False)

# Hide any unused subplots
for j in range(n_vars, n_rows * n_cols):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()

# Print summaries

rows = []
for s in summaries:
    var, rest = s.split(":", 1)
    m = re.search(r"bin is ([0-9\.]+\s*to\s*[0-9\.]+) with (\d+)", rest)
    if m:
        bin_range, count = m.group(1), int(m.group(2))
    else:
        bin_range, count = rest.strip(), None
    rows.append({"Variable": var, "BinRange": bin_range, "Count": count})

df_summary = pd.DataFrame(rows)

# 2) Style it: make the BinRange text navy and the Count text dark red
styled = (
    df_summary
    .style
    .set_properties(subset=["BinRange"], **{"color": "skyblue", "font-weight": "bold"})
    .set_properties(subset=["Count"],    **{"color": "red"})
    .set_table_styles([
        {"selector": "th", "props": [("font-size","16px"), ("text-align","left")]},
        {"selector": "td", "props": [("padding","6px"), ("text-align","left")]}
    ])
)

display(styled)
```

```python
#Q–Q Plots for Continuous Variables
from scipy import stats

display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Q–Q Plots for Continuous Variables</h2>'))
# Determine grid size (e.g. 3 columns)
n = len(cont_vars)
cols = 3
rows = int(np.ceil(n/cols))

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*4))
axes = axes.flatten()

for ax, var in zip(axes, cont_vars):
    data = df_all[var].dropna()
    # Q–Q plot
    stats.probplot(data, dist="norm", plot=ax)
    ax.get_lines()[1].set_color('red')       # make the fit line red
    ax.set_title(f"Q–Q Plot of {var}")
    ax.tick_params(axis='both', which='major', labelsize=10)
    # tighten the axis range to the 1st/99th percentile
    lo, hi = np.percentile(data, [1, 99])
    ax.set_xlim(stats.norm.ppf(0.01), stats.norm.ppf(0.99))
    ax.set_ylim(lo, hi)
    # improve marker style
    for line in ax.get_lines():
        line.set_marker('.')
        line.set_markersize(4)
        line.set_alpha(0.6)

# hide any empty subplots
for ax in axes[n:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

### Interpretation of data normality

1. Shapiro–Wilk on the full data  
   - Every continuous variable gave Shapiro–Wilk p < 0.05. In a sample of ~60 000, the test flags even tiny tail departures as “significant,” so this tells us only that our distributions are all not normal

2. Histograms & Q–Q plots (1st–99th percentile zoom)  
   - Haemoglobin, Heart Rate, Cholesterol, Age – Visually bell-shaped in the centre, Q–Q points hug the line for most of the range. A few extreme tails remain, but the middle 95–98 % looks approximately normal  
   - CRP, PHQ-9, SleepHours – Even after dropping the extreme 1 % on each side, the histograms stay skewed and the Q–Q plots bend away from the line—clear evidence of heavy tails or floor effects (e.g. many zeros in PHQ-9).

3. Conclusion  
   1. Variables with “trivial deviations” (Hb, heart rate, cholesterol) – Safe to treat as approximately normal in most regression or ANOVA models.  
   2. Variables that remain non-normal (CRP, PHQ-9, SleepHours) – Either transform (e.g. log-transform CRP) or use non-parametric methods (e.g. Spearman correlations, Wilcoxon tests) when strict distributional assumptions are needed

```python
# Define biomarker thresholds
biomarker_thresholds = {
    'SleepHours': {'min': None, 'max': 6, 'or_min': 9, 'or_max': None}, # <6 or >9
    'BPXPLS': {'min': None, 'max': 60, 'or_min': 100, 'or_max': None}, # <60 or >100
    'CRP': {'min': 3, 'max': None}, # >3
    'LBXTC': {'min': None, 'max': 125, 'or_min': 200, 'or_max': None}, # ~125-200 normal, so <125 or >200 abnormal
    'LBXHGB': {'male_min': None, 'male_max': 13.5, 'female_min': None, 'female_max': 12.0} # <13.5 (men), <12.0 (women)
}
# Initialize a dictionary to store results
abnormal_percentages = {}

# Get the total number of sampled persons
total_sampled = len(df_all)

# Iterate through each biomarker and its thresholds
for biomarker, thresholds in biomarker_thresholds.items():
    if biomarker == 'LBXHGB':
        # Handle gender-specific thresholds for Haemoglobin
        male_abnormal_count = df_all[
            (df_all['RIAGENDR'] == 1) &
            (df_all[biomarker] < thresholds['male_max'])
        ].shape[0]
        female_abnormal_count = df_all[
            (df_all['RIAGENDR'] == 2) &
            (df_all[biomarker] < thresholds['female_max'])
        ].shape[0]
        abnormal_count = male_abnormal_count + female_abnormal_count
    else:
        # Handle other biomarkers with min/max/or_min/or_max thresholds
        condition = pd.Series(False, index=df_all.index)

        if thresholds.get('min') is not None:
            condition = condition | (df_all[biomarker] > thresholds['min'])
        if thresholds.get('max') is not None:
            condition = condition | (df_all[biomarker] < thresholds['max'])
        if thresholds.get('or_min') is not None:
            condition = condition | (df_all[biomarker] > thresholds['or_min'])
        if thresholds.get('or_max') is not None:
            condition = condition | (df_all[biomarker] < thresholds['or_max'])

        # Ensure we only count non-missing values
        abnormal_count = df_all[condition & df_all[biomarker].notna()].shape[0]

    # Calculate percentage and store
    abnormal_percentage = (abnormal_count / total_sampled) * 100
    abnormal_percentages[biomarker] = abnormal_percentage

display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Percentage of Sampled Persons with Abnormal Biomarker Levels</h2>'))
# Create a DataFrame for display
abnormal_df = pd.DataFrame(list(abnormal_percentages.items()), columns=['Biomarker', 'Percentage Abnormal (%)'])
display(abnormal_df)

fig, ax = plt.subplots(figsize=(6,4))

# simple bar plot
bars = ax.bar(
    abnormal_df['Biomarker'],
    abnormal_df['Percentage Abnormal (%)'],
    edgecolor='black',
    linewidth=0.8
)

# add value labels
for bar in bars:
    h = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        h + 0.5,
        f"{h:.1f}%",
        ha='center',
        va='bottom',
        fontsize=10
    )

# fix ticks and labels in one go
ax.set_xticks(range(len(abnormal_df)))
ax.set_xticklabels(abnormal_df['Biomarker'], rotation=45, ha='right')

# styling
ax.set_title('Percent of Sample with Abnormal Biomarker Levels')
ax.set_ylabel('Percentage Abnormal (%)')
ax.set_ylim(0, abnormal_df['Percentage Abnormal (%)'].max() * 1.2)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
