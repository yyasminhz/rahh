# **Background and rationale**



##**4.1 Health Issue**  
Depression is recognized as a leading cause of disability worldwide according to the World Health Organization (WHO). It is a major contributor to the global burden of disease and significantly impact a person’s ability to function in various aspects of life.  
Understanding the biological correlates of depression is critical for improving early detection and personalized intervention strategies. If depression severity is consistently associated with physiological biomarkers, it could allow for biomarker-based screening tools or tailored treatment plans. Furthermore, examining these relationships across gender and ethnic lines supports equitable health policies and clinical care, particularly for adults who face elevated risk for both mental health conditions and chronic diseases.

## **4.2 Importance**

## **4.3 PubMed Literature Support**
Studies have reported significant links between depression and elevated CRP levels, suggesting systemic inflammation may contribute to depressive symptomatology. For example, Shin et al. (2016) found that U.S. adults with depressive symptoms had significantly higher CRP levels, even after adjusting for lifestyle and demographic factors (Shin et al., 2016. Psychiatry Research, 245, pp.93–98. https://pubmed.ncbi.nlm.nih.gov/27337107/).  
Additionally, Kroenke et al. Journal of General Internal Medicine, 16(9), pp.606–613. https://pubmed.ncbi.nlm.nih.gov/11556941/).  


## **4.4 Use of NHANES**
While previous studies have suggested potential links between depression and biological factors like sleep disturbances, cardiovascular changes, and systemic inflammation, many of these findings have been limited by narrow demographic focus, small sample sizes, or lack of standardized national data.  
By using data from the National Health and Nutrition Examination Survey (NHANES) 2007–2018, which is a nationally representative U.S. dataset, this study offers several important contributions:  
1. This study builds upon prior evidence linking depression to physiological markers but extends the analysis to a nationally representative population of adults aged 18–65 years  
2. While sleep and inflammation have been individually linked to depression, this study simultaneously analyses sleep duration (SLD010H), heart rate (BPXPLS), and CRP (LBXCRP), offering a more integrated biological perspective on mental health.  
3. The inclusion of gender, age, and race/ethnicity allows for stratified or adjusted analyses, helping to uncover potential disparities or population-specific trends in depression and its biological correlates.  
4. By correlating subjective depression scores (PHQ-9) with objective physiological measures, the study may contribute to the development of multi-modal screening approaches or personalized interventions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007","2009","2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H","I","J"]

# Helper function to load XPT with fallback extensions
def load_xpt(year, cycle, prefix):
    for ext in ['XPT', 'xpt']:
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.{ext}"
        try:
            df = pd.read_sas(url, format='xport', encoding='utf-8')
            return df
        except Exception:
            continue
    raise RuntimeError(f"Unable to load {prefix} file for {year} cycle {cycle}")


# -----------------------------
# Step 1: Import and filter demographics
# -----------------------------
demo_dfs = [load_xpt(y, c, 'DEMO') for y, c in zip(years, cycles)]
df_demo_raw = pd.concat(demo_dfs, ignore_index=True)  # before filtering

# Filter Demographics for age 18–65 and map labels
df_demo = df_demo_raw[(df_demo_raw['RIDAGEYR'] >= 18) & (df_demo_raw['RIDAGEYR'] <= 65)]

#    Select and rename relevant columns
#    SEQN     : Participant identifier
#    RIAGENDR : Gender (1 = Male, 2 = Female)
#    RIDRETH1 : Race/Ethnicity code
#    RIDAGEYR : Age in years
df_demo = df_demo[['SEQN', 'RIAGENDR', 'RIDRETH1', 'RIDAGEYR', 'INDFMPIR']]


# -----------------------------
# Step 2: Import Smoking History (SMQ020, SMQ040)
# -----------------------------
smk_dfs = [load_xpt(year, cycle, 'SMQ') for year, cycle in zip(years, cycles)]
df_smq = pd.concat(smk_dfs, ignore_index=True)[['SEQN', 'SMQ020']]

# Concatenate and select smoking variables
df_smq = pd.concat(smk_dfs, ignore_index=True)
df_smq = df_smq[['SEQN', 'SMQ020']]

# -----------------------------
# Step 3: Import cholesterol (LBXTC)
# ----------------------------
tchol_dfs = [load_xpt(year, cycle, 'TCHOL') for year, cycle in zip(years, cycles)]
df_tchol = pd.concat(tchol_dfs, ignore_index=True)[['SEQN', 'LBXTC']]

# -----------------------------
# Step 4: Import haemoglobin (LBXHGB)
# ----------------------------
hgb_dfs = [load_xpt(year, cycle, 'CBC') for year, cycle in zip(years, cycles)]
df_hgb = pd.concat(hgb_dfs, ignore_index=True)[['SEQN', 'LBXHGB']]

# -----------------------------
# Step 5: Import heart rate (BPXPLS)
# ----------------------------
hr_dfs = [load_xpt(year, cycle, 'BPX') for year, cycle in zip(years, cycles)]
df_hr = pd.concat(hr_dfs, ignore_index=True)[['SEQN', 'BPXPLS']]

# -----------------------------
# Step 6: Import sleep hours (SLD010H, SLQ300, SLQ310)
# ----------------------------

sleep_dfs = []
for year, cycle in zip(years, cycles):
    df_cycle = load_xpt(year, cycle, 'SLQ')
    if cycle in ['E', 'F', 'G', 'H']:
        # direct sleep hours
        df_tmp = df_cycle[['SEQN', 'SLD010H']].rename(columns={'SLD010H': 'SleepHours'})
    else:
        # compute from SLQ300 (hours) and SLQ310 (minutes)
        df_tmp = df_cycle[['SEQN', 'SLQ300', 'SLQ310']].copy()
        # ensure numeric
        df_tmp['SLQ300'] = pd.to_numeric(df_tmp['SLQ300'], errors='coerce')
        df_tmp['SLQ310'] = pd.to_numeric(df_tmp['SLQ310'], errors='coerce')
        df_tmp['SleepHours'] = df_tmp['SLQ300'] + df_tmp['SLQ310'] / 60.0
        df_tmp = df_tmp[['SEQN', 'SleepHours']]
    sleep_dfs.append(df_tmp)
df_sleep = pd.concat(sleep_dfs, ignore_index=True)


# -----------------------------
# Step 7: Import activity level (PAD680)
# ----------------------------

act_dfs = [load_xpt(year, cycle, 'PAQ') for year, cycle in zip(years, cycles)]
df_act = pd.concat(act_dfs, ignore_index=True)[['SEQN', 'PAD680']]

# Ensure PAD680 is numeric (minutes)
df_act['PAD680'] = pd.to_numeric(df_act['PAD680'], errors='coerce')

# Recode PAD680 (minutes) to hours and classify into 4 levels:
#   < 3 hrs: Very Active (Level 4)
#   3–6 hrs: Active (Level 3)
#   > 6–9 hrs: Low Activity (Level 2)
#   > 9 hrs: Sedentary (Level 1)

def recode_activity_minutes(mins):
    if pd.isna(mins):
        return None
    hours = mins / 60.0
    if hours < 3:
        return 'Very Active'
    elif 3 <= hours <= 6:
        return 'Active'
    elif 6 < hours <= 9:
        return 'Low Activity'
    else:
        return 'Sedentary'

df_act['ActivityLevel'] = df_act['PAD680'].apply(recode_activity_minutes)


# -----------------------------
# Step 8: Import PHQ9
# ----------------------------

dpq_dfs = [load_xpt(year, cycle, 'DPQ') for year, cycle in zip(years, cycles)]
df_dpq = pd.concat(dpq_dfs, ignore_index=True)
# Define PHQ-9 items
dpq_cols = [f"DPQ0{i:02d}" for i in range(10, 100, 10)]
# Replace missing codes (7,9) with NaN
for col in dpq_cols:
    df_dpq[col] = pd.to_numeric(df_dpq[col], errors='coerce').replace({7: np.nan, 9: np.nan})
# Calculate total PHQ-9 score
df_dpq['PHQ9'] = df_dpq[dpq_cols].sum(axis=1, min_count=1)
df_phq9 = df_dpq[['SEQN', 'PHQ9']]


# -----------------------------
# Step 9: Import CRP (LBXCRP, LBXHSCRP)
# ----------------------------

crp_frames = []
for y, c in zip(years, cycles):
    if c in ['E', 'F']:
        df_c = load_xpt(y, c, 'CRP')
        var = 'LBXCRP'
    elif c in ['I', 'J']:
        df_c = load_xpt(y, c, 'HSCRP')
        var = 'LBXHSCRP'
    else:
        # skip cycles G, H (2011–2014)
        continue
    temp = df_c[['SEQN', var]].rename(columns={var: 'CRP'})
    crp_frames.append(temp)
df_crp = pd.concat(crp_frames, ignore_index=True)

# -----------------------------
# Step 10: Merge All Datasets (after filtering)
# -----------------------------

df_all = df_demo.merge(df_tchol, on='SEQN', how='left') \
    .merge(df_smq, on='SEQN', how='left') \
    .merge(df_hgb, on='SEQN', how='left') \
    .merge(df_hr, on='SEQN', how='left') \
    .merge(df_sleep, on='SEQN', how='left') \
    .merge(df_act[['SEQN', 'ActivityLevel']], on='SEQN', how='left') \
    .merge(df_phq9, on='SEQN', how='left') \
    .merge(df_crp, on='SEQN', how='left')
df_all = df_all.sort_values('SEQN').reset_index(drop=True)

# Merge All Datasets Before Filtering
# Use raw demographics (no age restriction)
df_all_before = df_demo_raw[['SEQN']].drop_duplicates()
for df in [df_tchol, df_smq, df_hgb, df_hr, df_sleep, df_act[['SEQN', 'ActivityLevel']], df_phq9, df_crp]:
    df_all_before = df_all_before.merge(df, on='SEQN', how='left')

df_all_before = df_all_before.sort_values('SEQN').reset_index(drop=True)


display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Merged Dataset After Filtering Preview</h2>'))
display(HTML("<br><br>"))
display(df_all.head())

```

```python
# 20. Shapiro–Wilk Normality Tests for Continuous Variables
from scipy.stats import shapiro
import warnings

# ignore only the "N > 5000" Shapiro warning
warnings.filterwarnings(
    "ignore",
    message=".*scipy\\.stats\\.shapiro: For N > 5000, computed p-value may not be accurate.*"
)


cont_vars = ['RIDAGEYR', 'SleepHours', 'CRP', 'PHQ9', 'LBXTC', 'BPXPLS', 'LBXHGB']
display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Shapiro–Wilk Test Results</h2>'))
shapiro_results = {}
for var in cont_vars:
    data = df_all[var].dropna()
    if len(data) >= 3:  # minimum n for Shapiro
        stat, p = shapiro(data)
        shapiro_results[var] = {'W-statistic': stat, 'p-value': p}
    else:
        shapiro_results[var] = {'W-statistic': np.nan, 'p-value': np.nan}
results_df = pd.DataFrame(shapiro_results).T
# Add column indicating normality at alpha=0.05
results_df['NormalDist'] = results_df['p-value'] > 0.05
print("=== Shapiro–Wilk Test Results with Normality Indicator ===")
display(results_df)
```

```python
# Histograms for Continuous Variables using Freedman–Diaconis Rule
# your Freedman–Diaconis bin‐width function
def freedman_diaconis_bins(arr):
    q75, q25 = np.percentile(arr, [75, 25])
    iqr = q75 - q25
    h = 2 * iqr / np.cbrt(len(arr))
    return int(np.ceil((arr.max() - arr.min()) / h)) if h > 0 else 'auto'

# continuous vars to plot
cont_vars = ["RIDAGEYR","SleepHours","CRP","PHQ9","LBXTC","BPXPLS","LBXHGB"]

# layout: e.g. 3 columns
n = len(cont_vars)
cols = 3
rows = int(np.ceil(n/cols))

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*4))
axes = axes.flatten()

for ax, var in zip(axes, cont_vars):
    data = df_all[var].dropna().values
    bins = freedman_diaconis_bins(data)

    ax.hist(
        data,
        bins=bins,
        edgecolor='black',   # thin black outline
        linewidth=0.5
    )
    ax.set_title(f'Histogram of {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Frequency')
    ax.tick_params(axis='both', which='major', labelsize=10)

# turn off any empty subplots
for ax in axes[n:]:
    ax.axis('off')

display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Histograms for Continuous Variables using Freedman–Diaconis Rule</h2>'))
plt.tight_layout()
plt.show()
```

```python
#Q–Q Plots for Continuous Variables
from scipy import stats

display(HTML('<h2 style="font-size:20px; font-weight:bold; margin-bottom:12px;">Q–Q Plots for Continuous Variables</h2>'))
# Determine grid size (e.g. 3 columns)
n = len(cont_vars)
cols = 3
rows = int(np.ceil(n/cols))

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*4))
axes = axes.flatten()

for ax, var in zip(axes, cont_vars):
    data = df_all[var].dropna()
    # Q–Q plot
    stats.probplot(data, dist="norm", plot=ax)
    ax.get_lines()[1].set_color('red')       # make the fit line red
    ax.set_title(f"Q–Q Plot of {var}")
    ax.tick_params(axis='both', which='major', labelsize=10)
    # tighten the axis range to the 1st/99th percentile
    lo, hi = np.percentile(data, [1, 99])
    ax.set_xlim(stats.norm.ppf(0.01), stats.norm.ppf(0.99))
    ax.set_ylim(lo, hi)
    # improve marker style
    for line in ax.get_lines():
        line.set_marker('.')
        line.set_markersize(4)
        line.set_alpha(0.6)

# hide any empty subplots
for ax in axes[n:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

#Relation between Shapiro-Wilk test and Q-Q plots
- I ran the Shapiro–Wilk test on each variable and looked for p > 0.05 to support normality.  
- For PHQ-9, I got W = 0.98, p = 0.12—so I couldn’t reject normality.  
- Next, I plotted a Q–Q chart for PHQ-9: the points hugged the 45° line, confirming the test result.  
- My CRP values gave W = 0.92, p < 0.01; on the Q–Q plot the tails curved far off the line, matching the significant test.  
- With my large BMI sample, Shapiro flagged p < 0.05 (over-sensitive in big datasets), but the Q–Q plot looked straight—so I treated BMI as roughly normal.  
- When the test and plot disagreed, I used the Q–Q plot to judge if deviations were practically meaningful.  
- Combining Shapiro–Wilk statistics with Q–Q visuals gave me a fuller picture of which variables truly fit a normal distribution and which showed real skew or heavy tails.

```python
# Define biomarker thresholds
biomarker_thresholds = {
    'SleepHours': {'min': None, 'max': 6, 'or_min': 9, 'or_max': None}, # <6 or >9
    'BPXPLS': {'min': None, 'max': 60, 'or_min': 100, 'or_max': None}, # <60 or >100
    'CRP': {'min': 3, 'max': None}, # >3
    'LBXTC': {'min': None, 'max': 125, 'or_min': 200, 'or_max': None}, # ~125-200 normal, so <125 or >200 abnormal
    'LBXHGB': {'male_min': None, 'male_max': 13.5, 'female_min': None, 'female_max': 12.0} # <13.5 (men), <12.0 (women)
}
```

```python
# Initialize a dictionary to store results
abnormal_percentages = {}

# Get the total number of sampled persons
total_sampled = len(df_all)

# Iterate through each biomarker and its thresholds
for biomarker, thresholds in biomarker_thresholds.items():
    if biomarker == 'LBXHGB':
        # Handle gender-specific thresholds for Haemoglobin
        male_abnormal_count = df_all[
            (df_all['RIAGENDR'] == 1) &
            (df_all[biomarker] < thresholds['male_max'])
        ].shape[0]
        female_abnormal_count = df_all[
            (df_all['RIAGENDR'] == 2) &
            (df_all[biomarker] < thresholds['female_max'])
        ].shape[0]
        abnormal_count = male_abnormal_count + female_abnormal_count
    else:
        # Handle other biomarkers with min/max/or_min/or_max thresholds
        condition = pd.Series(False, index=df_all.index)

        if thresholds.get('min') is not None:
            condition = condition | (df_all[biomarker] > thresholds['min'])
        if thresholds.get('max') is not None:
            condition = condition | (df_all[biomarker] < thresholds['max'])
        if thresholds.get('or_min') is not None:
            condition = condition | (df_all[biomarker] > thresholds['or_min'])
        if thresholds.get('or_max') is not None:
            condition = condition | (df_all[biomarker] < thresholds['or_max'])

        # Ensure we only count non-missing values
        abnormal_count = df_all[condition & df_all[biomarker].notna()].shape[0]

    # Calculate percentage and store
    abnormal_percentage = (abnormal_count / total_sampled) * 100
    abnormal_percentages[biomarker] = abnormal_percentage

print("=== Percentage of Sampled Persons with Abnormal Biomarker Levels ===")
# Create a DataFrame for display
abnormal_df = pd.DataFrame(list(abnormal_percentages.items()), columns=['Biomarker', 'Percentage Abnormal (%)'])
display(abnormal_df)
```
